```python
import pandas as pd
pd.options.display.max_rows = 50
```

```python
mgb_nodes = pd.read_csv("ljo_exams/data/Mangabo Nodes.csv", delimiter = '\t')
mgb_exams_nodes_mgb_id = pd.read_csv("ljo_exams/data/Mangabo mgb_id Exam Degrees to 1609.csv", delimiter = '\t')
```

```python
# Man'gabo node list

mgb_exams_nodes = []

for index, row in mgb_exams_nodes_mgb_id.iterrows():
    nodes = mgb_nodes[mgb_nodes['mgb_id'] == row['mgb_exams_nodes_mgb_id']]
    for index, row in nodes.iterrows():
        item = [row['mgb_id'], row['ljo_id'], row['name'], row['ancestral_seat'], row['choronym']]
        print(item)
        mgb_exams_nodes.append(item)

df = pd.DataFrame(mgb_exams_nodes, columns=['mgb_id', 'ljo_id', 'name', 'ancestral_seat', 'choronym'])
df.drop_duplicates()
df.to_csv('ljo_exams/neo4j/neo4j_mgb_exams_nodes.csv', encoding='utf-8', sep='\t', index=False)
```

```python
# Man'gabo edge list
mgb_exam_id = pd.read_csv("ljo_exams/data/Mangabo ljo_id Exam Degrees to 1609.csv", delimiter = '\t')
mgb_nodes = pd.read_csv("ljo_exams/data/Mangabo Nodes.csv", delimiter = '\t')
mgb_edges = pd.read_csv("ljo_exams/data/Mangabo Edges.csv", delimiter = '\t')
```

```python
mgb_kin_edgelist = []

# Iterate through the ids of exam passers identified by Lee Jaeok
for index, row in mgb_exam_id.iterrows():
    nodes = mgb_nodes[mgb_nodes['ljo_id'] == row['mgb_exam_id']]


    for index, row in nodes.iterrows():

        # Degree 1: Ego appears in source
        edges = mgb_edges[mgb_edges['source'] == row['mgb_id']]
        for index, row in edges.iterrows():
            item = [row['source'], row['target'], row['relationship']]
            print(item)
            mgb_kin_edgelist.append(item)
            
            # Degree 2: Ego's sources' sources
            edges = mgb_edges[mgb_edges['target'] == row['source']]
            for index, row in edges.iterrows():
                item = [row['source'], row['target'], row['relationship']]
                print(item)
                mgb_kin_edgelist.append(item)
                
                # Degree 3: Ego's sources' sources' sources
                edges = mgb_edges[mgb_edges['target'] == row['source']]
                for index, row in edges.iterrows():
                    item = [row['source'], row['target'], row['relationship']]
                    print(item)
                    mgb_kin_edgelist.append(item)
```

```python
# Iterate through the ids of exam passers identified by Lee Jaeok
for index, row in mgb_exam_id.iterrows():
    nodes = mgb_nodes[mgb_nodes['ljo_id'] == row['mgb_exam_id']]

    for index, row in nodes.iterrows():

        # Degree 1: Ego appears in targets
        edges = mgb_edges[mgb_edges['target'] == row['mgb_id']]
        for index, row in edges.iterrows():
            item = [row['source'], row['target'], row['relationship']]
            print(item)
            mgb_kin_edgelist.append(item)
           
            # Degree 2: Ego's targets' targets
            edges = mgb_edges[mgb_edges['source'] == row['target']]
            for index, row in edges.iterrows():
                item = [row['source'], row['target'], row['relationship']]
                print(item)
                mgb_kin_edgelist.append(item)
               
                # Degree 3: Ego's targets' targets' targets
                edges = mgb_edges[mgb_edges['source'] == row['target']]
                for index, row in edges.iterrows():
                    item = [row['source'], row['target'], row['relationship']]
                    print(item)
                    mgb_kin_edgelist.append(item)

df = pd.DataFrame(mgb_kin_edgelist, columns=['source', 'target', 'relationship'])
df.drop_duplicates()
df.to_csv('ljo_exams/neo4j/neo4j_mgb_edgelist.csv', encoding='utf-8', sep='\t', index=False)
```
