```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
row = []

for i in range(1259):
    j = i + 1
    if j < 10:
        j = '000' + str(j)
    elif j < 100:
        j = '00' + str(j)
    elif j < 1000:
        j = '0' + str(j)
    else:
        j = str(j)

    url = 'https://db.itkc.or.kr/dir/nodeViewPrint?grpId=hj&itemId=MI&dataId=ITKC_MI_'+ j + 'A'
    html = urlopen(url)
    soup = BeautifulSoup(html, 'html.parser')
    
    # id
    id = 'ITKC_MI_' + j + 'A'

    # title
    title = soup.find('div', {'class': 'list_tit2 list_tit3'})
    title = title.find('h3')
    title = title.text

    # name
    name = soup.find('th', text='성명')
    name = name.parent
    name = name.find('td')
    name = name.text

    # birthyear
    birthyear = soup.find('th', text='생년')
    birthyear = birthyear.parent
    birthyear = birthyear.find('td')
    birthyear = birthyear.text

    # deathyear
    deathyear = soup.find('th', text='몰년')
    deathyear = deathyear.parent
    deathyear = deathyear.find('td')
    deathyear = deathyear.text

    # choronym
    choronym = soup.find('th', text='본관')
    choronym = choronym.parent
    choronym = choronym.find('td')
    choronym = choronym.text

    row.append([id, title, name, birthyear, deathyear, choronym])
    
    print(id)
    time.sleep(1)

df = pd.DataFrame(row, columns=['id', 'title', 'name', 'birthyear', 'deathyear', 'choronym'])
df.to_csv('munjip.csv', encoding='utf-8', sep='\t', index=False)
```
